```python
from elasticsearch import Elasticsearch

# 数据来源:data/example_book_data.jsonl
es = Elasticsearch([{'host': 'localhost', 'port': 9200}])
```

```python
body = {
    "query": {
        "match": {
            # 查询内容"美术"分词(ik_max_word)为: ["美术"]
            "name": {
                "query": "美术",
                "boost": 1
            }
        }
    }
}
'''
Returns documents that match a provided text, number, date or boolean value. The provided text is analyzed before matching.

query
    (Required) Text, number, boolean value or date you wish to find in the provided <field>.

    The match query analyzes any provided text before performing a search. This means the match query can search text fields for analyzed tokens rather than an exact term.
'''
'''
GET test/_search
`body`
'''
es.search(index='test', body=body)  # 9条
```

```python
es.search(index='test', body={
    "query": {
        "match": {
            # 查询内容"计算"分词(ik_max_word)为: ["计算"]
            "name": {
                "query": "计算"
            }
        }
    }
})  # 46条
```

```python
es.search(index='test', body={
    "query": {
        "match": {
            # 查询内容"美术计算"分词(ik_max_word)为: ["美术", "计算"]
            "name": {
                "query": "美术计算"
            }
        }
    }
})  # 55条
```

```python
'''
operator
    (Optional, string) Boolean logic used to interpret text in the query value. Valid values are:

    OR (Default)
        For example, a query value of capital of Hungary is interpreted as capital OR of OR Hungary.
    AND
        For example, a query value of capital of Hungary is interpreted as capital AND of AND Hungary.
'''
es.search(index='test', body={
    "query": {
        "match": {
            "name": {
                "query": "美术计算机",
                "operator": "OR"  # 文档索引条件:[查询内容分词结果] 交 [text字段分词结果] != [];即[查询内容分词结果]至少有一个元素属于[text字段分词结果]
            }
        }
    }
})
```

```python
es.search(index='test', body={
    "query": {
        "match": {
            "name": {
                "query": "美术计算机",
                "operator": "AND"  # 索引条件:[查询内容分词结果] in [text字段分词结果]
            }
        }
    }
})
```

```python
'''
minimum_should_match
    (Optional, string) Minimum number of clauses that must match for a document to be returned.

    See the minimum_should_match parameter(https://www.elastic.co/guide/en/elasticsearch/reference/7.11/query-dsl-minimum-should-match.html) for valid values and more information.
'''
es.search(index='test', body={
    "query": {
        "match": {
            "name": {
                # 查询内容"数据库教程"分词(ik_max_word)为: ["数据库", "数据", "库", "教程"]
                "query": "规范教材",
                "minimum_should_match": "1"  # 索引条件:len([查询内容分词结果] 交 [text字段分词结果]) >= 1
            }
        }
    }
})
```

```python
'''
fuzziness
    (Optional, string) Maximum edit distance allowed for matching. See Fuzziness(https://www.elastic.co/guide/en/elasticsearch/reference/7.11/common-options.html#fuzziness) for valid values and more information.
'''
'''
分析:
"教育部规划教材：建筑工程预算（第3版）"的分词结果为:['教育部', '教育', '部', '规划', '教材', '建筑工程', '建筑', '工程预算', '工程', '预算', '第', '3', '版']

"预算"与"计算"的编辑距离为1===>满足"fuzziness": 1(即允许匹配的最大编辑距离为1)
'''
es.search(index='test', body={
    "query": {
        "match": {
            "name": {
                "query": "计算",
                "fuzziness": "1"
            }
        }
    },
    "sort": [{"_score": "asc"}],
    "_source": ["name"],
})
```
