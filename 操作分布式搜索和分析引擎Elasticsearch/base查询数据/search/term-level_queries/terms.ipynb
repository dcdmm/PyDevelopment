```python
from elasticsearch import Elasticsearch

# 数据来源:data/example_book_data.jsonl
es = Elasticsearch([{'host': 'localhost', 'port': 9200}])
```

```python
body = {
    "query": {
        "terms": {
            "name": ["十一五", "教育"],
            "boost": 1
        }
    }
}
# he terms query is the same as the term query, except you can search for multiple values.
'''
GET test/_search
`body`
'''
es.search(index='test', body=body)

```
